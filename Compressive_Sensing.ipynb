```python
# Parameters
N = 100  # Length of the signal
sparsity = 0.1  # Ratio of nonzero elements
M = 50  # Number of measurements

# Generate sparse signal
sparse_signal = np.zeros(N)
num_nonzero = int(sparsity * N)
nonzero_indices = np.random.choice(N, num_nonzero, replace=False)
sparse_signal[nonzero_indices] = np.random.randn(num_nonzero)

# Create a sensing matrix (random Gaussian matrix)
phi = np.random.randn(M, N)

# Perform compressive sensing (measurements)
y = phi @ sparse_signal

# Reconstruct the sparse signal using pseudo-inverse (basic recovery method)
reconstructed_signal = np.linalg.pinv(phi) @ y

# Plot the original and reconstructed sparse signals
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.stem(sparse_signal)
plt.title("Original Sparse Signal")
plt.xlabel("Index")
plt.ylabel("Amplitude")
plt.grid()

plt.subplot(2, 1, 2)
plt.stem(reconstructed_signal)
plt.title("Reconstructed Signal")
plt.xlabel("Index")
plt.ylabel("Amplitude")
plt.grid()

plt.tight_layout()
plt.show()
```

```python
# Parameters
N = 100  # Length of the signal
sparsity = 0.1  # Ratio of nonzero elements
M = 50  # Number of measurements

# Generate sparse signal
sparse_signal = np.zeros(N)
num_nonzero = int(sparsity * N)
nonzero_indices = np.random.choice(N, num_nonzero, replace=False)
sparse_signal[nonzero_indices] = np.random.randn(num_nonzero)

# Create a sensing matrix (random Gaussian matrix)
phi = np.random.randn(M, N)

# Perform compressive sensing (measurements)
y = phi @ sparse_signal

# Reconstruct the sparse signal using pseudo-inverse (basic recovery method)
reconstructed_signal = np.linalg.pinv(phi) @ y

# Plot settings
plt.figure(figsize=(12, 6))
plt.style.use("ggplot")  # Using a default available style

# Plot the original sparse signal
plt.subplot(2, 1, 1)
plt.plot(sparse_signal, linestyle='-', marker='o', markersize=5, label='Original Signal', color='b', alpha=0.7)
plt.title("Original Sparse Signal", fontsize=14, fontweight='bold')
plt.xlabel("Index", fontsize=12)
plt.ylabel("Amplitude", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Plot the reconstructed signal
plt.subplot(2, 1, 2)
plt.plot(reconstructed_signal, linestyle='-', marker='s', markersize=5, label='Reconstructed Signal', color='r', alpha=0.7)
plt.title("Reconstructed Signal", fontsize=14, fontweight='bold')
plt.xlabel("Index", fontsize=12)
plt.ylabel("Amplitude", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Improve layout
plt.tight_layout()
plt.show()
```
