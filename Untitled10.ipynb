```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
n = np.arange(10)
ph1 = 0
ph2_list = [np.pi/2, np.pi]

for ph2 in ph2_list:
    x = np.sin(2 * np.pi * 0.1 * n + ph1)
    y = np.sin(2 * np.pi * 0.1 * n + ph2)

    rxy = np.correlate(x, y, mode='full')
    nr = np.arange(-len(n) + 1, len(n))

    maxR = np.max(rxy)
    indR = np.argmax(rxy)
    lag_at_max_corr = nr[indR]

    rxx = np.correlate(x, x, mode='full')
    energy_x = rxx[len(n) - 1]

    ryx = np.correlate(y, x, mode='full')

    print(f'\nFor phase shift ph2 = {ph2} ({"cosine" if ph2 == np.pi/2 else "Ï€"})')
    print(f'The correlation at lag zero is: {rxy[np.where(nr == 0)][0]}.')
    print(f'The maximum correlation is at lag {lag_at_max_corr}.')
    print(f'Auto-correlation at lag zero: {energy_x}.')
    print(f'Cross-correlation does not satisfy commutativity: {not np.allclose(rxy, ryx[::-1])}')

    plt.figure(figsize=(8, 10))

    plt.subplot(3, 1, 1)
    plt.stem(n, x)
    plt.xlabel('Time index n')
    plt.ylabel('Amplitude')
    plt.title('Signal x(n)')
    plt.grid()

    plt.subplot(3, 1, 2)
    plt.stem(n, y)
    plt.xlabel('Time index n')
    plt.ylabel('Amplitude')
    plt.title(f'Signal y(n) with phase shift {ph2}')
    plt.grid()

    plt.subplot(3, 1, 3)
    plt.stem(nr, rxy)
    plt.xlabel('Lag index')
    plt.ylabel('Amplitude')
    plt.title('Cross-Correlation x[n] * y[n]')
    plt.grid()

    plt.tight_layout()
    plt.show()
```

```python
n = np.arange(10)
ph1 = 0
ph2_list = [np.pi/2, np.pi]

for ph2 in ph2_list:
    x = np.sin(2 * np.pi * 0.1 * n + ph1)
    y = np.sin(2 * np.pi * 0.1 * n + ph2)

    rxy = np.correlate(x, y, mode='full')
    nr = np.arange(-len(n) + 1, len(n))

    maxR = np.max(rxy)
    indR = np.argmax(rxy)
    lag_at_max_corr = nr[indR]

    rxx = np.correlate(x, x, mode='full')
    energy_x = rxx[len(n) - 1]

    ryx = np.correlate(y, x, mode='full')

    rxy_conv = np.convolve(x, y[::-1], mode='full')

    print(f'\nFor phase shift ph2 = {ph2} ({"cosine" if ph2 == np.pi/2 else "Ï€"})')
    print(f'The correlation at lag zero is: {rxy[np.where(nr == 0)][0]}.')
    print(f'The maximum correlation is at lag {lag_at_max_corr}.')
    print(f'Auto-correlation at lag zero: {energy_x}.')
    print(f'Cross-correlation does not satisfy commutativity: {not np.allclose(rxy, ryx[::-1])}')
    print(f'Cross-correlation using convolution matches: {np.allclose(rxy, rxy_conv)}')

    plt.figure(figsize=(8, 10))

    plt.subplot(3, 1, 1)
    plt.stem(n, x)
    plt.xlabel('Time index n')
    plt.ylabel('Amplitude')
    plt.title('Signal x(n)')
    plt.grid()

    plt.subplot(3, 1, 2)
    plt.stem(n, y)
    plt.xlabel('Time index n')
    plt.ylabel('Amplitude')
    plt.title(f'Signal y(n) with phase shift {ph2}')
    plt.grid()

    plt.subplot(3, 1, 3)
    plt.stem(nr, rxy)
    plt.xlabel('Lag index')
    plt.ylabel('Amplitude')
    plt.title('Cross-Correlation x[n] * y[n]')
    plt.grid()

    plt.tight_layout()
    plt.show()

csv_filename = "/content/motor_voltages.csv"
df = pd.read_csv(csv_filename)

voltage_A = df["Voltage A Min"].values
voltage_B = df["Voltage B Min"].values
voltage_C = df["Voltage C Min"].values

corr_AB = np.correlate(voltage_A, voltage_B, mode='full')
corr_AC = np.correlate(voltage_A, voltage_C, mode='full')
corr_BC = np.correlate(voltage_B, voltage_C, mode='full')

print(f"\nCorrelation between Voltage A and B: Max = {np.max(corr_AB)}")
print(f"Correlation between Voltage A and C: Max = {np.max(corr_AC)}")
print(f"Correlation between Voltage B and C: Max = {np.max(corr_BC)}")
```

```python
def compute_correlation(transmitted, received):
    correlation = np.correlate(received, transmitted, mode='full')
    max_index = np.argmax(correlation)
    estimated_shift = max_index - (len(transmitted) - 1)
    return correlation, estimated_shift

if __name__ == '__main__':
    transmitted = np.array([1, -1, 1, 1, -1, 1, -1, -1])
    shift_amount = int(input("Enter the shift amount: "))
    preamble_length = len(transmitted)

    received_clean = np.concatenate((np.zeros(shift_amount), transmitted, np.zeros(shift_amount)))
    noise = np.random.normal(0, 0.1, len(received_clean))
    received = received_clean + noise

    # Select the portion of the received signal from shift_amount to the end of the received signal
    selected_received = received[shift_amount:]

    correlation, estimated_shift = compute_correlation(transmitted, selected_received)
    print("Estimated shift (lag):", estimated_shift)

    # Update the lags array to match the length of correlation
    lags = np.arange(-len(transmitted) + 1, len(correlation) - len(transmitted) + 1)

    plt.figure(figsize=(16, 6))

    plt.subplot(2, 2, 1)
    plt.plot(received, label='Received Signal', marker='o')
    transmitted_indices = np.arange(len(transmitted))
    plt.plot(transmitted_indices, transmitted, label='Transmitted Preamble (Unshifted)', marker='x', linestyle='--', color='red')
    plt.axvline(x=estimated_shift, color='green', linestyle='--', label=f'Estimated Preamble Start in Received (lag={estimated_shift})')
    plt.title("Comparison of Received Signal & Unshifted Transmitted Preamble")
    plt.xlabel("Sample Index")
    plt.ylabel("Amplitude")
    plt.legend()

    plt.subplot(2, 2, 2)
    plt.plot(lags, correlation, label='Cross-Correlation', marker='o')
    plt.title('Cross-Correlation')
    plt.xlabel('Lag (samples)')
    plt.ylabel('Correlation')
    plt.axvline(estimated_shift, color='red', linestyle='--', label=f'Estimated Lag: {estimated_shift} samples')
    plt.legend()

    # Plot the selected portion of the received signal
    plt.subplot(2, 2, 3)
    plt.plot(selected_received, label='Selected Received Signal (From Shifted Position)', marker='o', color='purple')
    transmitted_indices = np.arange(len(transmitted))
    plt.plot(transmitted_indices, transmitted, label='Transmitted Preamble (Unshifted)', marker='x', linestyle='--', color='red')
    plt.axvline(x=estimated_shift, color='green', linestyle='--', label=f'Estimated Preamble Start in Received (lag={estimated_shift})')
    plt.title(f'Selected Received Signal (Shifted by {shift_amount} samples)')
    plt.xlabel('Sample Index')
    plt.ylabel('Amplitude')
    plt.legend()

    plt.tight_layout()
    plt.show()
```
